```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.stats import zscore

# Imputasi missing values dengan rata-rata
df['MarketCap'] = df['MarketCap'].fillna(df['MarketCap'].mean())
df['LastPrice'] = df['LastPrice'].fillna(df['LastPrice'].mean())

# Menghapus baris yang masih memiliki nilai NaN setelah imputasi
df.dropna(inplace=True)

# Menstandarkan data Kapitalisasi dan Harga Terakhir
scaler = StandardScaler()
df[['MarketCap', 'LastPrice']] = scaler.fit_transform(df[['MarketCap', 'LastPrice']])

# Menggunakan metode Z-Score untuk mendeteksi outliers
z_scores = zscore(df[['MarketCap', 'LastPrice']])
df_no_outliers = df[(z_scores < 3).all(axis=1)]  # Hapus data dengan Z-Score lebih dari 3

# Mengambil sampel acak dari data yang sudah diproses
df_random_sample = df_no_outliers.sample(n=10, random_state=42)  # Mengambil 10 baris acak

# Menampilkan data acak
df_random_sample

```

```python
from sklearn.cluster import KMeans

# Menentukan jumlah cluster (misalnya 3)
kmeans = KMeans(n_clusters=3, random_state=42)

# Menentukan cluster
df['Cluster'] = kmeans.fit_predict(df[['MarketCap', 'LastPrice']])

# Menampilkan hasil clustering
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['MarketCap'], y=df['LastPrice'], hue=df['Cluster'], palette='viridis')
plt.title('Hasil Clustering Saham')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Menghitung Silhouette Score
sil_score = silhouette_score(df[['MarketCap', 'LastPrice']], df['Cluster'])
print(f'Silhouette Score: {sil_score}')
```
